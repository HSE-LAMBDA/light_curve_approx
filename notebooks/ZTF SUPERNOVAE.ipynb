# Скачиваем реальные кривые блеска сверхновых с ZTF
https://sites.astro.caltech.edu/ztf/bts/bts.php

```python
import pandas as pd
from ztfquery import lightcurve
import time
import numpy as np
from tqdm import tqdm


path = "https://sites.astro.caltech.edu/ztf/bts/explorer.php?f=s&subsample=trans&classstring=&classexclude=&quality=y&purity=y&ztflink=lasair&lastdet=&startsavedate=&startpeakdate=&startra=&startdec=&startz=&startdur=&startrise=&startfade=&startpeakmag=&startabsmag=&starthostabs=&starthostcol=&startb=&startav=&endsavedate=&endpeakdate=&endra=&enddec=&endz=&enddur=&endrise=&endfade=&endpeakmag=19.0&endabsmag=&endhostabs=&endhostcol=&endb=&endav=&reverse=y&format=csv"
df = pd.read_csv(path, sep =',')
```

Для каждого объекта указан его тип, в дальнейшем будем сохранять в файл номер_в_этой_таблице.csv датафрейм для каждого объекта из таблицы. Таким образом есть биекция по номерам и можно быстро разметить.

### Экваториальные координаты (RA, Dec в таблице) в другом формате, нежели поддерживаемый формат для базы ZTF:

```python
for alph in df['Dec'].values:
    print(alph.split(':'))
```

Поэтому конвертируем в требуемый формат:

```python
def convert_coordinates(dataframe):
    ra = dataframe['RA'].values
    dec = dataframe['Dec'].values
    convert_ra = []
    convert_dec = []
    
    for alph, declin in zip(ra, dec):
        alpha = alph.split(':')
        new_alpha = int(alpha[0]) * 15 + int(alpha[1]) * 15 / 60 + float(alpha[2]) * 15 / 3600 # convert to degrees
        convert_ra.append(new_alpha)
        declination = declin.split(':')
        assert int(declination[0]) is not None
        
        if int(declination[0]) < 0:
            new_declination = abs(int(declination[0])) + int(declination[1]) / 60 + float(declination[2]) / 3600
            convert_dec.append(-new_declination)
            
        elif int(declination[0]) >= 0:
            assert int(declination[0]) >= 0
            new_declination = int(declination[0]) + int(declination[1]) / 60 + float(declination[2]) / 3600
            convert_dec.append(new_declination)
    return convert_ra, convert_dec
        
```

```python
start_time = time.time()
dframes_lc = []
ras, decs = convert_coordinates(df)
#print(ras, decs)
count_of_empty_df = 0
for idx, (ra, dec) in tqdm(enumerate(zip(ras, decs))):
    lcq = lightcurve.LCQuery().query_position(ra, dec, 1)
    df_lc = lcq.data
    if df_lc.index.values.shape[0] != 0:
        dframes_lc.append(df_lc)
        df_lc.to_csv("/home/masha/AstroAugumetation/DataFrames/{}.csv".format(idx))
    else:
        count_of_empty_df += 1
end = time.time() - start_time
print(end) # примерное время скачивания 2531 датафреймов - 7 часов
print(len(dframes_lc), count_of_empty_df)
```
